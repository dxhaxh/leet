```python
#combinations

#given n and k, return all combinations(order does not matter) of k integers from the range [1, ..., n]

class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        def backtrack(first, currentList):
            if len(currentList)==k:
                output.append(currentList[:])
                return
            for i in range(first, n+1):
                currentList.append(i)
                backtrack(i+1, currentList)
                currentList.pop()
        
        output=[]
        backtrack(1, [])
        return output
```

```python
#n queens

l=[ [0 for i in range(0, n)] for j in range(0, n) ]
res=[]
def getConfig(l):
    config=[]
    for i in range(0, n):
        curRow=''
        for j in range(0, n):
            if l[i][j]==0:
                curRow+='.'
            else:
                curRow+='Q'
        config.append(curRow)
    return config
def tryPlace(r, c, l):
    for i in range(0, n):
        for j in range(0, n):
            if i==r and l[i][j]==1:
                return False
            if j==c and l[i][j]==1:
                return False
            if abs(i-r)==abs(j-c) and l[i][j]==1:
                return False
    return True
def dfs(r, c, l):
    if r==n:
        #append result to res
        #means we have placed n queens without issue
        res.append(getConfig(l))
        return
    if c==n:
        #backtrack
        return
    if tryPlace(r, c, l):
        l[r][c]=1
        dfs(r+1, 0, l)

    l[r][c]=0
    dfs(r, c+1, l)
dfs(0, 0, l)
return res
```

```python
#number of islands

#solved with dfs, visit an island then dfs marking visited with 'X'
#if you come across a 1 then you know you came across a new island since previously visited islands will already be
#marked by 'X' from dfs, because dfs marks the whole island with 'X' once it visits an island
#so if you come across a 1 then increment res variable for num of islands and call dfs on it

def dfs(i, j, grid):
    if i>=len(grid) or j>=len(grid[0]) or i<0 or j<0 or grid[i][j] == 'X' or grid[i][j] == '0':
        return
    grid[i][j] = 'X'
    dfs(i, j+1, grid)
    dfs(i+1, j, grid)
    dfs(i, j-1, grid)
    dfs(i-1, j, grid)
res=0
for i in range(0, len(grid)):
    for j in range(0, len(grid[0])):
        if grid[i][j]=='1':
            res+=1
            dfs(i, j, grid)
return res
```

```python
#mouse jumping on stones problem

#given an array stones, stones[i] is the value of the ith stone
#mouse starts on index 0 and jumps from stone to stone
#value of a jump from index i to index j is (j-i)*stone[j], ie length of jump * value of landing stone
#after summing all jumps mouse has score
#maximize this score

#this could be done with a dp table in O(n^2) time
#but notice that we can start from end and just jump to next highest stone and calculate like that in O(n) time
#for ex in example below start at end at 1. each inverse jump from 1 counts as 1, ie length of inverse jump * value of cur gives points to cur
#with every extra jump length after 10 we are giving up 10 points for 1, therefore make cur 10 and continue from 10
#till we find better than 10 worth sacrificing for

nums = [3, 5, 4, 7, 2, 12, 8, 10, 1]
n = len(nums)

prev = n - 1
res = prev * nums[prev]

for i in range(n - 2, -1, -1):
    if(nums[i] > nums[prev]):
        res += i * (nums[i] - nums[prev])    # i * nums[i]  - i * nums[prev]
        prev = i
        
print(res)
```

```python
#min stack

#implement a stack that supports push, pop, top, getMin operations
#push: push element on the stack
#pop: remove element on top of stack(without returning it)
#top: get element on top of stack
#getMin: get the value of minimum item on stack

#all these operations must run in O(1) time

#all regular stack operations are easy to have in O(1) time
#to make getMin run in O(1) time, we make each node in the stack have an attribue that is the 2nd value in the tuple
#of each element on stack. This value corresponds to min so far.
#So for any item (x, y) on the stack, we know that y is the minimum value of any item added before x was added on stack

class MinStack:

    def __init__(self):
        self.stack=[('dummy val', 2**31-1)]

    def push(self, val: int) -> None:
        if val<self.stack[-1][1]:
            self.stack.append((val, val))
        else:
            self.stack.append((val, self.stack[-1][1]))

    def pop(self) -> None:
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1][0]

    def getMin(self) -> int:
        return self.stack[-1][1]
```

```python
#generate parenthesis

#given a number n, generate all posible valid chains of n pairs of valid parentheses

#example: n=1, res=['()']
#example: n=3, res=["((()))","(()())","(())()","()(())","()()()"]

#solution uses backtracking, builds a tree of possible parenthesis

stack=[]
res=[]
def backtrack(openCount, closeCount):
    if openCount==closeCount and openCount==n:
        res.append(''.join(stack))
        return
    if openCount<n:
        stack.append('(')
        backtrack(openCount+1, closeCount)
        stack.pop()
    if closeCount<openCount:
        stack.append(')')
        backtrack(openCount, closeCount+1)
        stack.pop()

backtrack(0, 0)
return res
```

```python
#implement trie (prefix tree)

class Trie:

    def __init__(self):
        self.children={}
        self.endOfWord=True

    def insert(self, word: str) -> None:
        cur=self
        for letter in word:
            if letter in cur.children:
                cur=cur.children[letter]
            else:
                cur.children[letter] = Trie()
                cur=cur.children[letter]
                cur.endOfWord=False
        cur.endOfWord=True

    def search(self, word: str) -> bool:
        cur=self
        for letter in word:
            if not letter in cur.children:
                return False
            else:
                cur=cur.children[letter]
        return cur.endOfWord
        

    def startsWith(self, prefix: str) -> bool:
        cur=self
        for letter in prefix:
            if not letter in cur.children:
                return False
            else:
                cur=cur.children[letter]
        return True


# Your Trie object will be instantiated and called as such:
# obj = Trie()
# obj.insert(word)
# param_2 = obj.search(word)
# param_3 = obj.startsWith(prefix)
```

```python
#reorder list (reorder linked list)

#given a linked list of the form L0 -> L1 -> L2 -> L3 -> ... -> Ln-1 -> Ln
#reorder the linked list to be of form L0 -> Ln -> L1 -> Ln-1 -> 2 -> Ln-2 -> ...
#return the head of the new list.
#Do not modify the node values only modify the connections of the nodes

#this solution runs in O(n) time and O(n) memory. 
#first scan through the linked list completely and put node i at index i in an array l
#then notice that for each i(i<n//2) we connect i with n-i(n is length of list) and then n-i is connected to i+1
#do this for each i
#then modify the last index n//2 such that its next node is None for it to be the end of the list


# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        l=[]
        cur=head
        while cur!=None:
            l.append(cur)
            cur = cur.next
        cur=head
        for i in range(0, len(l)//2):
            l[i].next=l[len(l)-1-i]
            l[len(l)-1-i].next=l[i+1]
        l[len(l)//2].next=None
        return head
```

```python
#copy list with random pointer 

#given a linked list where in addition to the next pointer, each node also has a random pointer
#where node.random points to some random node in the list or to null
#create a deepcopy of this linked list and return the head of this deepcopy

"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""

class Solution:
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        if head==None:  #in case there are no nodes in linked list given
            return None
        d={}
        l=[]
        deepCopyl=[]
        cur=head
        while cur!=None:
            l.append(cur)
            d[id(cur)]=len(l)-1
            deepCopyl.append(Node(cur.val))
            cur=cur.next
        for i in range(0, len(l)):
            originalNode=l[i]
            copyNode=deepCopyl[i]

            if i==len(l)-1:
                copyNode.next=None
            else:
                copyNode.next=deepCopyl[i+1]
            
            if originalNode.random==None:
                copyNode.random=None
            else:
                randomIndex=d[id(originalNode.random)]
                copyNode.random=deepCopyl[randomIndex]
        return deepCopyl[0]
```

```python
#linked list cycle

#given the head of a linked list, determined if there is a cylce in the linked list
#return True if there is a cycle, False otherwise

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        cur=head
        visited=set()
        while cur!=None:
            if id(cur) in visited:
                return True
            visited.add(id(cur))
            cur=cur.next
        return False
```

```python
#lru cache

#Design a data structure that follows the constraints of a Least Recently Used (LRU) cache.
#Implement the LRUCache class:
#LRUCache(int capacity) Initialize the LRU cache with positive size capacity.
#int get(int key) Return the value of the key if the key exists, otherwise return -1.
#void put(int key, int value) Update the value of the key if the key exists. Otherwise, add the key-value pair to the cache. If the number of keys exceeds the capacity from this operation, evict the least recently used key.
#The functions get and put must each run in O(1) average time complexity.


class LRUCache:

    def __init__(self, capacity: int):
        self.capacity=capacity
        self.d={}
        self.next={}
        self.prev={}

    def get(self, key: int) -> int:
        if not key in self.d:
            return -1
        self.prev[self.next[key]]=self.prev[key]
        self.next[self.prev[key]]=self.next[key]
        self.prev[key] = self.prev['end']
        self.next[self.prev['end']] = key
        self.prev['end'] = key
        self.next[key] = 'end'
        return self.d[key]

    def put(self, key: int, value: int) -> None:
        if len(self.d)==0:   #nothing in yet so build the prev and next dicts
            self.d[key] = value
            self.next['begin'] = key
            self.next[key]='end'
            self.prev[key]='begin'
            self.prev['end'] = key
            return
        if key in self.d:
            self.d[key] = value
            self.prev[self.next[key]]=self.prev[key]
            self.next[self.prev[key]]=self.next[key]
            self.prev[key] = self.prev['end']
            self.next[self.prev['end']] = key
            self.prev['end'] = key
            self.next[key] = 'end'
            return
        if len(self.d)==self.capacity:
            toRemove=self.next['begin']
            self.next['begin'] = self.next[toRemove]
            self.prev[self.next[toRemove]] = 'begin'
            del self.next[toRemove]
            del self.prev[toRemove]
            del self.d[toRemove]
            self.d[key] = value
            self.prev[key] = self.prev['end']
            self.next[self.prev['end']] = key
            self.prev['end']=key
            self.next[key] = 'end'
            return
        self.d[key]=value
        self.prev[key] = self.prev['end']
        self.next[self.prev['end']] = key
        self.prev['end']=key
        self.next[key] = 'end'


# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```

```python
#find the duplicate number

#given a list of n+1 integers in the range [1,n],
#find the integer that is a duplicate
#note than a duplicate integer may be repeated more than 2 times
#i.e. [3,3,3,3,3] is a valid input to this problem,
#so we cannot solve it by summing list and then doing that sum minus n*(n+1)/2

#the way to do this problem is to notice that each integer in the list is in the range [1, n]
#there are a total of n+1 integers in the list so list indices are in the range [0, n]
#we treat the list integers sort of as a 'linked list of some sorts'
#each integer points us to the next index
#before moving on to the next index, we make the current one 0 (to signify that we visited it)
#we modify the list in place to use O(1) memory
#if we ever get to an index that is already 0, then we know this index is the duplicate integer
#since that means we have been pointed to this index before and made it 0
#we start at index 0 since we know we can never return to it because all the integers are in range [1, n]

#this solution is O(n) time complexity and uses O(1) extra memory

def findDuplicate(self, nums: List[int]) -> int:
    cur=0
    while nums[cur]!=0:
        prev=cur
        cur=nums[cur]
        nums[prev]=0
    return cur
```

```python
#invert binary tree

#given root of a binary tree, invert it and return its root
#inverting given any node, switch its right and left children, and that not just for the root
#but for every node in the tree

#solution uses a recursive function to do the action on a node and then to each of its children

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if root==None:
            return root
        if root.left==None and root.right==None:
            return root
        temp=root.right
        root.right=root.left
        root.left=temp
        self.invertTree(root.right)
        self.invertTree(root.left)
        return root
```

```python
#maximum depth of binary tree

#given root of binary tree, return its maximum depth

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if root==None:
            return 0
        return 1+max(self.maxDepth(root.right), self.maxDepth(root.left))
```

```python
#diameter of a binary tree

#given the root of a binary tree, return the diameter of it
#the diameter is the largest path from one node to another possible in a tree
#note: the diameter path need not pass by the root

#to solve this we calculate the diameter for each root(the diameter passing by that root)
#note that the diameter for a given root is  the height or the left subree plus the height of the right subtree
#we then update a global variable keeping track of the max diameter seen so far
#the max diameter is then returned
#note that for a given root, the diameter passing by it is always the height of left subtree plus height of right

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        largestDiameter=[0]     #we make this a list for it to be used as a global variable

        def height(root):
            if root==None:
                return 0
            leftHeight=height(root.left)
            rightHeight=height(root.right)
            diameter=leftHeight+rightHeight #diameter for this particular node

            largestDiameter[0] = max(largestDiameter[0], diameter) #note if instead of largestDiameter[0] we just updated largestDiameter without it being a list, then it would be local and there would be a largestDiameter variable for each height function called that is in the call stack and this would cause an error

            return 1+max(leftHeight, rightHeight)

        height(root)
        return largestDiameter[0]
```

```python
#check if binary tree is balanced

#A height-balanced binary tree is a binary tree in which the depth of the two subtrees (left and right) 
#of every node never differs by more than one.

def isBalanced(self, root: Optional[TreeNode]) -> bool:
    res=[True]
    def height(root):
        if root==None:
            return 0
        leftHeight=height(root.left)
        rightHeight=height(root.right)

        if abs(leftHeight-rightHeight)>=2:
            res[0]=False

        return 1+max(leftHeight, rightHeight)

    height(root)
    return res[0]
```

```python
#checking if two binary trees are the same

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        def same(p, q):
            if p==None and q==None:
                return True
            if p==None or q==None: #we only enter here if one is None and other isn't
                return False      #because the case both None is checked above
            if p.val!=q.val:
                return False
            return True and same(p.left, q.left) and same(p.right, q.right)
        
        return same(p, q)
```

```python
#subtree of another tree

#Given the roots of two binary trees root and subRoot, 
#return true if there is a subtree of root with the same structure and node values of subRoot and false otherwise.
#A subtree of a binary tree tree is a tree that consists of a node in tree and all of this node's descendants. 
#The tree tree could also be considered as a subtree of itself.

#the solution uses the solution of another leetcode problem has a helper function
#at each node, of the tree given by root, we check if that node and subroot form the same tree


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSubtree(self, root: Optional[TreeNode], subRoot: Optional[TreeNode]) -> bool:
        def same(p, q):
            if p==None and q==None:
                return True
            if p==None or q==None: #we only enter here if one is None and other isn't
                return False      #because the case both None is checked above
            if p.val!=q.val:
                return False
            return True and same(p.left, q.left) and same(p.right, q.right)

        def lookForSame(p, q):
            if p==None:
                return same(p, q)
            return same(p, q) or lookForSame(p.left, q) or lookForSame(p.right, q)
        
        return lookForSame(root, subRoot)
```

```python
#lowest common ancestor of a binary search tree

#given the root of a bst and nodes p, q, return the lowest common ancestor of p and q
#the tree is guaranteed to have at least 2 nodes in it
#p is not equal to q
#all nodes have unique values
#a node is the lowest common ancestor of itself by definition(ie it q is a child of p, then p is the lowest common ancestor)

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if root==p or root==q:
            return root
        if (root.val>p.val and root.val<q.val) or (root.val>q.val and root.val<p.val):
            return root
        if (root.val>p.val and root.val>q.val):
            return self.lowestCommonAncestor(root.left, p, q)
        else:
            return self.lowestCommonAncestor(root.right, p, q)
```

```python
#binary tree level order traversal

#Given the root of a binary tree, return the level order traversal of its nodes' values. 
#(i.e., from left to right, level by level)

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        res=[]
        def level(root, l):
            if root==None:
                return
            if len(res)<l:
                res.append([])
            res[l-1].append(root.val)
            level(root.left, l+1)
            level(root.right, l+1)

        level(root, 1)
        return res
```

```python
#binary tree right side view

#Given the root of a binary tree, imagine yourself standing on the right side of it, 
#return the values of the nodes you can see ordered from top to bottom.

#note that we cannot just go right and add that node to res, since if at some level there is no node to the right
#the the rightmost node of that level is returned even if that node is a left child of it's parent
#to solve this problem we use the solution from level order traversal of a binary tree
#the list listOfLevels will be a list of lists, where the ith list of listOfLevels is the ith level(0 indexed) nodes
#listed from left to right. If a node is None it does not appear in listOfLevels
#so then we simply just take the last entry of every list in listOfLevels to get the right side view of the tree


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        listOfLevels=[]
        def inOrderLevels(root, level):
            if root==None:
                return
            if len(listOfLevels)<level:
                listOfLevels.append([])
            listOfLevels[level-1].append(root.val)
            inOrderLevels(root.left, level+1)
            inOrderLevels(root.right, level+1)

        inOrderLevels(root, 1)        
        res=[]
        for l in listOfLevels:
            res.append(l[-1])
        return res
```

```python
#count good nodes in a binary tree

#Given a binary tree root, a node X in the tree is named good if in the path from root to X 
#there are no nodes with a value greater than X.
#Return the number of good nodes in the binary tree.


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def goodNodes(self, root: TreeNode) -> int:
        res=[0]
        def gn(root, prevMax):
            if root==None:
                return
            if root.val>=prevMax:
                res[0]+=1
                prevMax=root.val
            gn(root.left, prevMax)
            gn(root.right, prevMax)
        
        gn(root, -10**4-1)
        return res[0]
```

```python
#validate binary search tree

#given the root of a binary search tree, check if it is a valid binary search tree
#a binary search tree is valid if:
#1. The left subtree of a node contains only nodes with keys less than the node's key.
#2. The right subtree of a node contains only nodes with keys greater than the node's key.
#3. Both the left and right subtrees must also be binary search trees.

#the solution uses the fact that a binary tree that has a sorted in order traversal of node values is a valid BST.
#so do an in order traversal of the tree and check if by going from left to right in the in order traversal of the nodes,
#the values always increase, if so then tree is a valid BST, otherwise it is not.


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        inOrderNodes=[]
        def inOrderTraversal(root):
            if root==None:
                return
            inOrderTraversal(root.left)
            inOrderNodes.append(root.val)
            inOrderTraversal(root.right)
        
        inOrderTraversal(root)
        for i in range(1, len(inOrderNodes)):
            if inOrderNodes[i-1]>=inOrderNodes[i]:
                return False
        return True
```

```python
#kth smallest element in a bst

#given the root of a bst and an integer k, return the kth smallest(0 indexed) element of the bst
#k will always be less or equal to the number of nodes in the tree

#the solution uses the fact that the in order traversal of a bst is sorted
#so just do an in order traversal and get the kth element of the list of nodes in the order obtained
#from the in order traversal
#note to make things a bit more efficient, we don't do the full in order traversal,
#we break of the function once we have visited at least k nodes while doing an in order traversal of the tree
#if we were not guranteed that k was less or equal to the number of nodes in the tree
#then we would do a full in order traversal, we would check if k is indeed less or equal to the number of nodes in the tree
#and then would return the kth smallest if k is indeed a valid value


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        kSmallest=[]
        def inOrderTraversal(root):
            if root==None:
                return
            if len(kSmallest)>=k:
                return
            inOrderTraversal(root.left)
            if len(kSmallest)>=k:
                return
            kSmallest.append(root.val)
            if len(kSmallest)>=k:
                return
            inOrderTraversal(root.right)
        
        inOrderTraversal(root)
        return kSmallest[k-1]
```

```python
#binary tree maximum path sum

#A path in a binary tree is a sequence of nodes where each pair of adjacent nodes in the sequence 
#has an edge connecting them. A node can only appear in the sequence at most once. 
#Note that the path does not need to pass through the root.
#The path sum of a path is the sum of the node's values in the path.
#Given the root of a binary tree, return the maximum path sum of any non-empty path.

#we are guaranteed to have at least one node in the tree
#the solution we use is a recursive bottom up solution
#for each node we calulate the maximum path sum that passes by that node
#node the maximum path sum passes by a node and both its left and right subrees only if it is rooted at that node
#we can use this if rooted result in calculating the max
#but the result the function returns will not be if it's rooted to not throw off
#any other prior result we are calculating


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        res=[-1001]
        def mps(root):
            #this defines a function that calculates the max
            #path sum on path starting at root
            if root==None:
                return 0
            rightPS=mps(root.right)
            leftPS=mps(root.left)
            ans=max(root.val, root.val+rightPS, root.val+leftPS)
            ansRooted=root.val+leftPS+rightPS
            res[0]=max(res[0], ans, ansRooted)
            return ans
        
        mps(root)
        return res[0]
```

```python
#last stone weight

#You are given an array of integers stones where stones[i] is the weight of the ith stone.
#We are playing a game with the stones. On each turn, we choose the heaviest two stones and smash them together. 
#Suppose the heaviest two stones have weights x and y with x <= y. The result of this smash is:
#If x == y, both stones are destroyed, and
#If x != y, the stone of weight x is destroyed, and the stone of weight y has new weight y - x.
#At the end of the game, there is at most one stone left.
#Return the weight of the last remaining stone. If there are no stones left, return 0.

#solution uses a max heap to keep track of the largest stone
#each turn(while there are still 2 or more stones in list), we pop 2 elements from max heap
#i.e. we take the two largest stones
#we then either add their difference(largest-smallest) back in heap or if they're equal do nothing and go next turn
#note heapq only supports min heap so we add - before pushing and after popping any element in the heap
#for heapq to be a max heap and not a min heap

import heapq
class Solution:
    def lastStoneWeight(self, stones: List[int]) -> int:
        #we want to build a max heap but heapq only supports min heaps
        #so make a new list of stones l
        #where the values are all negated
        #add negatives when popping/pushing accordingly
        l=[]
        for i in stones:
            l.append(-i)
        heapq.heapify(l)
        while len(l)>=2:
            y=-heapq.heappop(l)
            x=-heapq.heappop(l)
            if x==y:
                continue
            else:
                heapq.heappush(l, -(y-x))
        if len(l)==0:
            return 0
        return -l[0]
```

```python
#design twitter

#Design a simplified version of Twitter where users can post tweets, follow/unfollow another user, and is able to see the 10 most recent tweets in the user's news feed.
#Implement the Twitter class:
#Twitter() Initializes your twitter object.
#void postTweet(int userId, int tweetId) Composes a new tweet with ID tweetId by the user userId. Each call to this function will be made with a unique tweetId.
#List<Integer> getNewsFeed(int userId) Retrieves the 10 most recent tweet IDs in the user's news feed. Each item in the news feed must be posted by users who the user followed or by the user themself. Tweets must be ordered from most recent to least recent.
#void follow(int followerId, int followeeId) The user with ID followerId started following the user with ID followeeId.
#void unfollow(int followerId, int followeeId) The user with ID followerId started unfollowing the user with ID followeeId.

import heapq
class Twitter:

    def __init__(self):
        self.time=0
        self.feed={}#maps every userid to a list of all their (time, tweets)
        self.followers={}#maps a userid to a set of all their followings

    def postTweet(self, userId: int, tweetId: int) -> None:
        if not userId in self.feed:
            self.feed[userId]=[]
        self.feed[userId].append((self.time, tweetId))
        self.time+=1
        if len(self.feed[userId])>10:
            self.feed[userId].pop(0)
        if not userId in self.followers:
            self.followers[userId]=set()
        self.followers[userId].add(userId)

    def getNewsFeed(self, userId: int) -> List[int]:
        if not userId in self.followers:
            return []
        feed=[] #this will be a minheap of size 10(holds 10 latest tweets)
        for tweeter in self.followers[userId]:
            if not tweeter in self.feed: #if the user that is followed has no tweets
                continue
            for tweet in self.feed[tweeter]:
                heapq.heappush(feed, tweet)
                if len(feed)>10:
                    heapq.heappop(feed)
        feed.sort(key=lambda x: x[0], reverse=True)#sorted from largest to smallest
        res=[]
        for tweetTuple in feed:
            res.append(tweetTuple[1]) #to just get tweetId without time
        return res

    def follow(self, followerId: int, followeeId: int) -> None:
        if not followerId in self.followers:
            self.followers[followerId]=set()
        self.followers[followerId].add(followerId)
        self.followers[followerId].add(followeeId)

    def unfollow(self, followerId: int, followeeId: int) -> None:
        if (followerId in self.followers) and (followeeId in self.followers[followerId]):
            self.followers[followerId].remove(followeeId)


# Your Twitter object will be instantiated and called as such:
# obj = Twitter()
# obj.postTweet(userId,tweetId)
# param_2 = obj.getNewsFeed(userId)
# obj.follow(followerId,followeeId)
# obj.unfollow(followerId,followeeId)
```

```python
#combination sum 2

#Given a collection of candidate numbers (candidates) and a target number (target), 
#find all unique combinations in candidates where the candidate numbers sum to target.
#Each number in candidates may only be used once in the combination.
#Note: The solution set must not contain duplicate combinations.

class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        candidates.sort()
        res=[]
        sol=[]

        def backtrack(i, curSum):
            if curSum==target:
                res.append(sol[:]) #makes a copy of the sol and appends to res
                return
            if i>len(candidates)-1 or curSum>target:
                return

            for j in range(i, len(candidates)):
                if j>i and candidates[j]==candidates[j-1]:
                    continue
                sol.append(candidates[j])
                backtrack(j+1, curSum+candidates[j])
                sol.pop()
            


        backtrack(0, 0)
        return res
    
    
    
#problem can also be solved using a used(or banned) set
#this set is used to avoid duplicates, we first sort the array and then we check if the current index
#is an element that is a duplicate, the only way we use the current is if the previous was used
#if previous was not used, then we just go down branch where current is not used i.e. we never pick current
#and then return, otherwise we go down branch where it is picked and down branch where it is not picked

candidates.sort()
res=[]
cur=[]
used=set()

def backtrack(i, curSum):
    if curSum==target:
        res.append(cur[:])
        return
    if i==len(candidates) or curSum>target:
        return

    if i>0 and candidates[i]==candidates[i-1] and i-1 not in used:
        backtrack(i+1, curSum)
        return

    used.add(i)
    cur.append(candidates[i])
    backtrack(i+1, curSum+candidates[i])
    cur.pop()
    used.remove(i)

    backtrack(i+1, curSum)

backtrack(0, 0)
return res
```

```python
#letter combinations of a phone number

#Given a string containing digits from 2-9 inclusive, 
#return all possible letter combinations that the number could represent. Return the answer in any order

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        d={
            '2': ['a', 'b', 'c'],
            '3': ['d', 'e', 'f'],
            '4': ['g', 'h', 'i'],
            '5': ['j', 'k', 'l'],
            '6': ['m', 'n', 'o'],
            '7': ['p', 'q', 'r', 's'],
            '8': ['t', 'u', 'v'],
            '9': ['w', 'x', 'y', 'z']
        }
        if digits=='':
            return []
        if len(digits)==1:
            return d[digits]
        res=[]
        for letter in d[digits[0]]:
            for comb in self.letterCombinations(digits[1:]):
                res.append(letter+comb)
        
        return res
```

```python
#word search

#Given an m x n grid of characters board and a string word, return true if word exists in the grid.
#The word can be constructed from letters of sequentially adjacent cells, 
#where adjacent cells are horizontally or vertically neighboring. The same letter cell may not be used more than once.

class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        seen=set()
        ans=[False]
        
        def hunt(i, j, word):
            if word=='':
                ans[0] = ans[0] or True
                return
            if i>len(board)-1 or j>len(board[0])-1 or i<0 or j<0:
                return
            if (i, j) in seen:
                return
            if board[i][j]!=word[0]:
                return
            seen.add((i, j))
            hunt(i+1, j, word[1:])
            hunt(i-1, j, word[1:])
            hunt(i, j+1, word[1:])
            hunt(i, j-1, word[1:])
            seen.remove((i, j))

        for i in range(0, len(board)):
            for j in range(0, len(board[0])):
                if ans[0]:
                    return ans[0]
                hunt(i, j, word)
            
        return ans[0]
```

```python
#clone graph

#given a node of a connected undirected graph, create a deep copy of this graph and return the copy
#(return the copy node associated with the provided node)
#the graph contains no repeated edges(ie node only appears once at most in another nodes neighbors) and no self loops
#graph is conected

#note that in this graph, since it is undirected, it is a bit complicated to keep track of everything
#since if 1 and 4 are connected by an edge for example, 1 appears in 4.neighbors but 4 also appears in 1.neighbors
#to avoid any complications we approach this problem a bit differently
#we create two dictionaries, one called originalNodes which maps a node's(from original graph) id to that node
#and another called oldIdToNewNode which maps the original node's id to the copy node in the copy graph
#we then visit every node in the original graph and use originalNodes as a seen set
#after visiting each node, we map the id of that node to the node in originalNodes
#and we map the id of that node to a new node(its copy that we create on the spot) in oldIdToNewNode
#the we loop through the originalNodes dictionary and for every original node, we get every node in it's neighbors
#and append the associated copy neighbor to the neighbors list of the original node copy


"""
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
"""

from typing import Optional
class Solution:
    def cloneGraph(self, node: Optional['Node']) -> Optional['Node']:
        if node==None:
            return None
        originalNodes={} #maps original node id to node. needed to get the neighbours
        oldIdToNewNode={} #maps the old node Id to the new node that is it's copy in the new graph
        def visitNodes(node):#will visit all nodes in graph
            #we can use oldIdToNewNode dict as a seenDict
            if id(node) in originalNodes:
                return
            originalNodes[id(node)] = node
            oldIdToNewNode[id(node)] = Node(val = node.val)
            for child in node.neighbors:
                visitNodes(child)

        visitNodes(node)
        for originalNodeId in originalNodes:
            originalNode = originalNodes[originalNodeId]
            for child in originalNode.neighbors:
                oldIdToNewNode[originalNodeId].neighbors.append(oldIdToNewNode[id(child)])
        return oldIdToNewNode[id(node)]
```

```python
#walls and gates

#You are given a m x n 2D grid initialized with these three possible values.
# -1 - A wall or an obstacle.
# 0 - A gate.
# INF - Infinity means an empty room. We use the value 231 - 1 = 2147483647 to represent INF as you may assume that the distance to a gate is less than 2147483647.
#Fill each empty room with the distance to its nearest gate. If it is impossible to reach a gate, it should be filled with INF.

#solution uses a dfs solution, on each gate, we call dfs. If we encounted a wall return,
#otherwise we have a parameter c which represents the distance from the gate that dfs was originally called on
#if c is greater than current value of grid[i][j], then return cause we're not improving anything
#and wont improve anything from this point on since stuff from this point on is less than c
#because it was changed from another dfs from another gate
#if c is smaller than current grid[i][j] value, then make current grid[i][j] value equal to c
#and then call dfs on call adjacent squares with c value equal to c+1(since we get one square further from gate)

def sol(board):
    def dfs(i, j, c):
        if i>len(board)-1 or j>len(board[0])-1 or i<0 or j<0:
            return
        if board[i][j]==-1:
            return
        if board[i][j]!=0:
            if c<board[i][j]:
                board[i][j]=c
            else:
                return   #this makes so we dont need a seen set to keep visited nodes in
        dfs(i+1, j, c+1)
        dfs(i-1, j, c+1)
        dfs(i, j+1, c+1)
        dfs(i, j-1, c+1)
        
    for i in range(0, len(board)):
        for j in range(0, len(board[0])):
            if board[i][j]==0:
                dfs(i, j, 0)
    return board
```

```python
#rotting oranges

#You are given an m x n grid where each cell can have one of three values:
#0 representing an empty cell,
#1 representing a fresh orange, or
#2 representing a rotten orange.
#Every minute, any fresh orange that is 4-directionally adjacent to a rotten orange becomes rotten.
#Return the minimum number of minutes that must elapse until no cell has a fresh orange. If this is impossible, return -1.

#solution uses similar solution to walls and gates
#for each fresh orange we want want to know the distance to the nearest rotting orange
#this distance will correspond to the (minimum)minutes it takes for that fresh orange to turn rotten
#if an orange cannot get rotten then we leave it's value as 1.
#we create a function dfs that gets called on every rotten orange,
#it gets called with a distance attribute, c, which we will make 2 initially
#this is a tweak, we make the min distance 2 and then edit it later when we get the result
#if during dfs we go out of bounds, then return
#if we get to a 0 (i.e. no orange) then return
#if we get to 1, treat it as a special value and immediately make grid[i][j]=c
#if we get to any other value then check if c<grid[i][j] if it is then we found a better(closer to rotting) value
#for the orange at [i][j], else we return since we know that any orange from this point on was already touched by a closer
#rotting orange that we've already called dfs on
#note this works well because if we get to another 2(rotting orange) then c is never < 2 and so we return
#but for the one it is called at it works(and we want it to work on the 2 it is called on in order to explore other nodes)
#we then explore every adjacent node with c+1 since we get 1 further from the rotting orange
#at the end we check every node in the grid
#if there is a 1 then there is still a fresh orange in the grid so return -1
#otherwise get the maximum of the grid and subtract 2 to get the largest minimum distance of any fresh orange to rotting


class Solution:
    def orangesRotting(self, grid: List[List[int]]) -> int:
        def dfs(i, j, c):
            if i>len(grid)-1 or j>len(grid[0])-1 or i<0 or j<0:
                return
            if grid[i][j]==0:  #if empty cell then return
                return
            if grid[i][j]==1:
                grid[i][j]=c   #fresh orange, make it rot
            elif grid[i][j]>c:
                grid[i][j]=c  #if we found a better value(orange at i,j rots quicker) then make it this new better value
            elif c!=2:   #this is to not break out of the first dfs called on 2
                return  #but if we visit any other 2 then we return which is what we want
            dfs(i+1, j, c+1)  
            dfs(i-1, j, c+1)  #calls made with c+1 in because we get 1 further from rotten orange
            dfs(i, j+1, c+1)
            dfs(i, j-1, c+1)

        for i in range(0, len(grid)):
            for j in range(0, len(grid[0])):
                if grid[i][j]==2:
                    dfs(i, j, 2)  #if we see rotting orange, call dfs on it
        res=0
        for i in range(0, len(grid)):
            for j in range(0, len(grid[0])):
                if grid[i][j]==1: #means not all oranges rot, return -1
                    return -1
                res=max(res, grid[i][j]-2) #-2 since every call we made was with c starting at 2, so 2 more than actual distance
        return res
            
```

```python
#surrounded regions

#You are given an m x n matrix board containing letters 'X' and 'O', capture regions that are surrounded:
#Connect: A cell is connected to adjacent cells horizontally or vertically.
#Region: To form a region connect every 'O' cell.
#Surround: The region is surrounded with 'X' cells if you can connect the region with 'X' cells and none of the region cells are on the edge of the board.
#A surrounded region is captured by replacing all 'O's with 'X's in the input matrix board.

#this solution makes a dfs, this dfs is called on any node that is on edge of the grid and is 'O'
#we then visit all adjacent nodes keeping track of all the 'O' 's positions in a set
#this set contains all 'O' 's that do not get transformed
#we then check every possible (i, j) position in the grid, if this (i, j) is not in the O set then make it 'X'
#dfs is similar to the islands. Pretend the islands are formed of 'O' 's and we simply wanna get all the islands
#that have at least one part on an edge of the grid
 

class Solution:
    def solve(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        oSet=set()
        def dfs(i, j):
            if i>len(board)-1 or j>len(board[0])-1 or i<0 or j<0:
                return
            if board[i][j]=='X':
                return
            if (i, j) in oSet:
                return
            oSet.add((i, j))
            dfs(i+1, j)
            dfs(i-1, j)
            dfs(i, j+1)
            dfs(i, j-1)

        for i in range(0, len(board)):
            if board[i][0]=='O':
                dfs(i, 0)
            if board[i][len(board[0])-1]:
                dfs(i, len(board[0])-1)
        for j in range(0, len(board[0])):
            if board[0][j]:
                dfs(0, j)
            if board[len(board)-1][j]:
                dfs(len(board)-1, j)
        
        for i in range(0, len(board)):
            for j in range(0, len(board[0])):
                if (i, j) in oSet:
                    continue
                board[i][j] = 'X'
```

```python
#course schedule

#we are given an integer numCourses and an array prerequisites. The courses are numbered 0,1,...,numCourses-1
#the entries in prerequisites are 2-tuples of courses [x, y] which mean that we must take course y before taking course x
#We must take all courses. Return True if this is possible, False otherwise

#this problem basically boils down to a topological sort
#we start by building a (directed) graph where each node corresponds to a course
#and edge (y, x) is in the graph if we must take y before x
#we then simply must check if there is a cylce in this graph or not by using dfs
#we use dfs as seen in COMP251 where a node is white if undiscovered, gray while being discovered, and black when finished
#note if we're exploring and run into a gray node during dfs that mean this node is being discovered(and not finished)
#but the node we are at is a descendant of the node being discovered, thus there is a cycle in the graph
#and we change the global variable to indicate that there is a cycle
#note we don't want an ordering of courses so don't care about the topological sort
#we simply want if there is a cylce in the graph or not

class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        graph=[[] for i in range(0, numCourses)]#create a graph where graph[i] hold i's adjacency list
        for tup in prerequisites:
            x=tup[0]
            y=tup[1]
            graph[y].append(x)
        white=set()
        gray=set()
        black=set()
        res=[True]
        def dfs_visit(node):
            white.remove(node)
            gray.add(node)
            for child in graph[node]:
                if child in gray:
                    res[0]=False
                if child in white:
                    dfs_visit(child)
            gray.remove(node)
            black.add(node)
        
        for i in range(0, numCourses):
            white.add(i)
        for node in range(0, numCourses):
            if node in white:
                dfs_visit(node)
        return res[0]
```

```python
#course schedule 2

#similar to course schedule 1,
#we are given an integer numCourses and an array prerequisites. The courses are numbered 0,1,...,numCourses-1
#the entries in prerequisites are 2-tuples of courses [x, y] which mean that we must take course y before taking course x
#We must take all courses. return any possible ordering of courses or if impossible return empty array

#this problem is now asking for a topological sort
#we apply the topological sort algorithm from COMP251,
#run dfs and when a node is finished add it to the front of a linked list and return this linked list which
#will contain one possible topological sorting
#this works since descendants of any node are always finished before the parent(in a DAG)
#note however that topological sort is only possible for DAGs so we must also make sure there is no
#cycle in the graph, if there is a cylce then we return an empty array

class Solution:
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        graph=[[] for i in range(0, numCourses)]
        for tup in prerequisites:
            x=tup[0]
            y=tup[1]
            graph[y].append(x) #y must come before x, add x to y's adjacency list
        whiteSet=set()
        graySet=set()
        blackSet=set()
        acyclic=[True]
        res=[] #when a node is finished we add to front of list(append and at end reverse list is returned)
        def dfs_visit(node):
            whiteSet.remove(node)
            graySet.add(node)
            for child in graph[node]: #for child in node's adjacency list
                if child in whiteSet:
                    dfs_visit(child)
                if child in graySet:
                    acyclic[0] = False
            graySet.remove(node)
            blackSet.add(node)
            res.append(node)

        for node in range(0, numCourses):
            whiteSet.add(node)
        for node in range(0, numCourses):
            if node in whiteSet:
                dfs_visit(node)
        if not acyclic[0]:
            return []
        
        return res[::-1]
```

```python
#min cost to connect all points

#solution is a bit complicated since it implements a priority queue from scratch
#one problem with heapq is that we cannot change value of keys efficiently
#so I implemented this myself
#making a class PriorityQueue that has a heap and a dicitonary keeping track of what item is at what index
#that way we can easily change the key value of an item since we just get it's index and modify the value

class PriorityQueue:
    def __init__(self):
        self.heap=[] #heap elements will be of form [value, p], sorted by value
        self.indices={} #maps an element to it's index in the heap

    def minHeapify(self, i):
        A=self.heap
        left=2*i+1
        right=2*i+2
        smallest=i
        if left<len(A) and A[left][0]<A[i][0]:
            smallest=left
        if right<len(A) and A[right][0]<A[smallest][0]:
            smallest=right
        if smallest!=i:
            objectAtsmallest=self.heap[smallest][1]
            objectAti=self.heap[i][1]
            self.indicies[objectAti]=smallest
            self.indicies[objectAtsmallest] = i
            temp=A[smallest]
            A[smallest] = A[i]
            A[i] = temp
            self.minHeapify(smallest)

    def minHeapify2(self, i):
        A=self.heap
        left=2*i+1
        right=2*i+2
        smallest=i
        if left<len(A) and A[left][0]<A[i][0]:
            smallest=left
        if right<len(A) and A[right][0]<A[smallest][0]:
            smallest=right
        if smallest!=i:
            temp=A[smallest]
            A[smallest] = A[i]
            A[i] = temp

    def buildQueue(self, arr):
        self.heap=arr
        self.indicies={}
        for i in range((len(arr)//2), -1, -1):
            self.minHeapify2(i)
        for i in range(0, len(self.heap)):
            self.indicies[self.heap[i][1]]=i

    def extractMin(self):
        A=self.heap
        m=A[0]
        A[0] = A[-1]
        objectAtEnd=A[0][1]
        self.indicies[objectAtEnd]=0
        A.pop()
        del self.indicies[m[1]]
        self.minHeapify(0)
        return m[1]

    def decreaseKey(self, p, newValue):
        A=self.heap
        i=self.indicies[p]
        A[i][0] = newValue
        while i>0 and A[(i-1)//2][0]>A[i][0]:
            parentObject=A[(i-1)//2][1]
            self.indicies[parentObject] = i
            self.indicies[p] = (i-1)//2
            temp=A[(i-1)//2]
            A[(i-1)//2]=A[i]
            A[i]=temp
            i=(i-1)//2
            
    def __str__(self):
        return str(self.heap) + "\n" + str(self.indicies)


class Solution:
    def minCostConnectPoints(self, points: List[List[int]]) -> int:
        def dist(u, v):
            return abs(u[0]-v[0]) + abs(u[1]-v[1])
        
        t=0

        q=PriorityQueue()
        toMake=[[10**7, tuple(p)] for p in points]
        toMake[0][0]=0
        q.buildQueue(toMake)
        while len(q.heap)!=0:
            t+=q.heap[0][0]
            u=q.extractMin()
            print(u)
            for child in points:
                if tuple(child)==u:
                    continue
                if u==(7,0):
                    print(q.heap)
                    print(q.indicies)
                if tuple(child) in q.indicies and dist(u, child)<q.heap[q.indicies[tuple(child)]][0]:
                    q.decreaseKey(tuple(child), dist(u, child))
        return t
```

```python
#climbing stairs

#you are given n stairs. Every step you have the choice to either climb 1 stair or 2 stairs,
#in how many ways can you climb the n stairs

#this is the best solution using dp takes O(n) time and O(1) memory
#notice that to climb the nth stair, we either get to the n-1th stair and take one step
#or get to the n-2th stair and take 2 steps
#so its the number of ways to get to the n-1th stair plus the number of ways to get to the n-2th stair
#so the recursion is f(n)=f(n-1)+f(n-2), notice we only need previous 2 values so we can store those in variables
#instead of storing all values 1-n in an array which makes memory complexity better

if n==1:
    return 1
if n==2:
    return 2
a=1
b=2
for i in range(3, n+1):
    temp=a
    a=b
    b=b+temp
return b

#using an array, because array stores all values, this takes O(n) memory complexity
dp=[0, 1, 2]
if n in {0, 1, 2}:
return n
for i in range(3, n+1):
dp.append(dp[i-1]+dp[i-2])
return dp[-1]
```

```python
#decode ways

#you are given a string s which is just digits
#you are given the mapping '1' -> 'A', '2' -> 'B', ..., '26' -> 'Z'
#you want to know the number of ways to decode s
#example: s='11106' could be decoded as (11, 10, 6) or as (1, 1, 10, 6) so 2 ways
#note that 06 is not valid for mapping so (11, 1, 06) is not good and not counted as one of ways
#there could be no valid mapping

#we notice that we have the recurrence relation as follows
#f(s) = f(s[1:]) + f(s[2:]), if the first 2 chars of s form valid num <=26
#f(s) = f(s[1:]),           otherwise
#where f(s) is the number of ways to decode s
#we make a dp table and solve it from right to left
#of course we need to fill in the rightmost 2 entries correctly(this was the hardest part)
#then we just fill the table right to left and return dp[0] w

if s[0]=='0':
    return 0
if len(s)==1:
    return 1
dp=[0 for i in range(0, len(s))]

#fill in the last 2 entries of dp(we know there are at least 2 since len(s)>=2(checked above)) 
if s[-1]!='0':   #dp[-1] already set to 0, change only if s[-1]!=0
    dp[-1]=1
if s[-2]!='0':
    if s[-1]=='0':
        if int(s[-2]+s[-1])<=26:
            dp[-2]=1
        else:
            dp[-2]=0
    else:
        if int(s[-2]+s[-1])<=26:
            dp[-2]=2
        else:
            dp[-2]=1

#after filling in last 2 entries we have recurrence as follows
#f(s)=f(s[1:])+f(s[2:]) only if next 2 chars make int<=26
#else f(s)=f(s[1:])
for i in range(len(s)-3, -1, -1):
    if s[i]=='0':
        dp[i]=0
        continue
    if int(s[i] + s[i+1])<=26:
        dp[i] = dp[i+1] + dp[i+2]
    else:
        dp[i] = dp[i+1]
return dp[0]
```

```python
#unique paths

#given a nxm grid and starting at (0,0) you can only go either down or right and you cannot go out of bounds
#how many ways are there to get at (n-1, m-1)

#this solution could be done with dp using the recurrence relation f(i, j) = f(i+1, j)+f(i, j+1)
#but it can also be done another way. 
#Notice only moves are left or right and there are always a total of (m-1+n-1) moves being done m-1 of which are right
#and n-1 of which are down. The number of possible ways is the number of ways to order these m-1+n-1 moves
#m-1 of which are the same and n-1 of which are the same. So it's just (m-1+n-1) choose (m-1)

class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        def fact(n):
            if n==0:
                return 1
            t=1
            for i in range(1, n+1):
                t*=i
            return t
        def choose(n, k):
            return fact(n)/(fact(n-k)*fact(k))
        return int(choose(n-1+m-1, m-1))
```

```python
#min cost to connect all points

#You are given an array points representing integer coordinates of some points on a 2D-plane, where points[i] = [xi, yi].
#The cost of connecting two points [xi, yi] and [xj, yj] is the manhattan distance between them: 
#|xi - xj| + |yi - yj|, where |val| denotes the absolute value of val.
#Return the minimum cost to make all points connected. 
#All points are connected if there is exactly one simple path between any two points.

#this is just Prim's algorithm, find the mst for the given nodes(points on graph given) where edge weights between
#nodes are the manhattan distance between them

#in this solution we implement a min priority queue that supports the decrease key function from scratch
#this can also be done in another way by using heapq and a seen set and stopping once we have seen all nodes
#i.e. once all nodes are in the tree and thus connected

class PriorityQueue:
    def __init__(self):
        self.heap=[] #heap elements will be of form [value, p], sorted by value
        self.indices={} #maps an element to it's index in the heap

    def minHeapify(self, i):
        A=self.heap
        left=2*i+1
        right=2*i+2
        smallest=i
        if left<len(A) and A[left][0]<A[i][0]:
            smallest=left
        if right<len(A) and A[right][0]<A[smallest][0]:
            smallest=right
        if smallest!=i:
            objectAtsmallest=self.heap[smallest][1]
            objectAti=self.heap[i][1]
            self.indicies[objectAti]=smallest
            self.indicies[objectAtsmallest] = i
            temp=A[smallest]
            A[smallest] = A[i]
            A[i] = temp
            self.minHeapify(smallest)

    def minHeapify2(self, i):
        A=self.heap
        left=2*i+1
        right=2*i+2
        smallest=i
        if left<len(A) and A[left][0]<A[i][0]:
            smallest=left
        if right<len(A) and A[right][0]<A[smallest][0]:
            smallest=right
        if smallest!=i:
            temp=A[smallest]
            A[smallest] = A[i]
            A[i] = temp

    def buildQueue(self, arr):
        self.heap=arr
        self.indicies={}
        for i in range((len(arr)//2), -1, -1):
            self.minHeapify2(i)
        for i in range(0, len(self.heap)):
            self.indicies[self.heap[i][1]]=i

    def extractMin(self):
        A=self.heap
        m=A[0]
        A[0] = A[-1]
        objectAtEnd=A[0][1]
        self.indicies[objectAtEnd]=0
        A.pop()
        del self.indicies[m[1]]
        self.minHeapify(0)
        return m[1]

    def decreaseKey(self, p, newValue):
        A=self.heap
        i=self.indicies[p]
        A[i][0] = newValue
        while i>0 and A[(i-1)//2][0]>A[i][0]:
            parentObject=A[(i-1)//2][1]
            self.indicies[parentObject] = i
            self.indicies[p] = (i-1)//2
            temp=A[(i-1)//2]
            A[(i-1)//2]=A[i]
            A[i]=temp
            i=(i-1)//2
            
    def __str__(self):
        return str(self.heap) + "\n" + str(self.indicies)


class Solution:
    def minCostConnectPoints(self, points: List[List[int]]) -> int:
        def dist(u, v):
            return abs(u[0]-v[0]) + abs(u[1]-v[1])
        
        t=0

        q=PriorityQueue()
        toMake=[[10**7, tuple(p)] for p in points]
        toMake[0][0]=0
        q.buildQueue(toMake)
        while len(q.heap)!=0:
            t+=q.heap[0][0]
            u=q.extractMin()
            print(u)
            for child in points:
                if tuple(child)==u:
                    continue
                if u==(7,0):
                    print(q.heap)
                    print(q.indicies)
                if tuple(child) in q.indicies and dist(u, child)<q.heap[q.indicies[tuple(child)]][0]:
                    q.decreaseKey(tuple(child), dist(u, child))
        return t
```

```python
#swim in rising water

#You are given an n x n integer matrix grid where each value grid[i][j] represents the elevation at that point (i, j).
#The rain starts to fall. At time t, the depth of the water everywhere is t. 
#You can swim from a square to another 4-directionally adjacent square 
#if and only if the elevation of both squares individually are at most t. 
#You can swim infinite distances in zero time. Of course, you must stay within the boundaries of the grid during your swim.
#Return the least time until you can reach the bottom right square (n - 1, n - 1) if you start at the top left square (0, 0).

#solution uses a variation of bfs with a priority queue.

seen=set()
q=[(grid[0][0], (0, 0))]
while len(q)!=0:
    tup=heapq.heappop(q)
    curMax=tup[0]
    i, j = tup[1]
    if i==len(grid)-1 and j==len(grid[0])-1:
        return curMax                        #if we got to the destination return value
    if (i, j) in seen:
        continue
    seen.add((i, j))
    if i+1<len(grid) and not (i+1, j) in seen:
        toAdd=(max(grid[i+1][j], curMax), (i+1, j))       #to go to child we must be at time a least: max(curCell, childCell)
        heapq.heappush(q, toAdd)
    if i-1>=0 and not (i-1, j) in seen:
        toAdd=(max(grid[i-1][j], curMax), (i-1, j))
        heapq.heappush(q, toAdd)
    if j+1<len(grid[0]) and not (i, j+1) in seen:
        toAdd=(max(grid[i][j+1], curMax), (i, j+1))
        heapq.heappush(q, toAdd)
    if j-1>=0 and not (i, j-1) in seen:
        toAdd=(max(grid[i][j-1], curMax), (i, j-1))
        heapq.heappush(q, toAdd)
```

```python
#cheapest flight within k stops

#There are n cities connected by some number of flights. 
#You are given an array flights where flights[i] = [fromi, toi, pricei] indicates that there is a flight from city fromi 
#to city toi with cost pricei.
#You are also given three integers src, dst, and k, 
#return the cheapest price from src to dst with at most k stops. If there is no such route, return -1.

#variation of bfs, we have a limited depth k, we keep track of depth as well as distance
#if depth ever exceeds k, children are not pushed on heap

graph={}
connectsToDst=set()
for tup in flights:
    x=tup[0]
    y=tup[1]
    w=tup[2]
    if not x in graph:
        graph[x]=[]       #build the graph
    if not y in graph:
        graph[y]=[]
    if y==dst:
        connectsToDst.add(x)
    graph[x].append((y, w))

if len(connectsToDst)==0:
    return -1
q=[(0, 0, src)] #(price, num of stops, node)
while len(q)!=0:
    curPrice, stopsSoFar, node=heapq.heappop(q)
    if node==dst:
        return curPrice
    if stopsSoFar==k:
        if not node in connectsToDst:
            continue
        for child, weight in graph[node]:
            if child==dst:
                heapq.heappush(q, (curPrice+weight, stopsSoFar, child))
        continue
    if stopsSoFar<k:
        for child, weight in graph[node]:
            if child==dst:
                heapq.heappush(q, (curPrice+weight, stopsSoFar, child))
            else:
                heapq.heappush(q, (curPrice+weight, stopsSoFar+1, child))
return -1
```

```python
#maximum subarray

#Given an integer array nums, find the subarray with the largest sum, and return its sum.
#
#nums = [-2,1,-3,4,-1,2,1,-5,4], Output: 6

#for each point, we calculate the largest subarray sum ENDING at that point
#i.e. that point must be included as last
#we could store these values in a dp table where dp[i] is the largest
#subarray sum ending with index i, where index i must be included, but we 
#could make this o(1) memory by only keeping track of the sum of the current
#subarray we are calculating, and the largest sum seen so far out of all subarrays
cur=-10**5
res=-10**5  #intial values will be smaller than possible min
for i in nums:
    if cur<0: #in any case if cur is negative we'd rather have only one num, the current one we are at
        cur=i
        res=max(res, i)
    else:
        cur+=i
        res=max(res, cur)
return res
```

```python
#target sum

#You are given an integer array nums and an integer target.
#You want to build an expression out of nums by adding one of the symbols '+' and '-' before each integer in nums 
#and then concatenate all the integers.
#For example, if nums = [2, 1], you can add a '+' before 2 and a '-' before 1 and concatenate them to build the expression "+2-1".
#Return the number of different expressions that you can build, which evaluates to target

#0/1 knapsack problem but we have to use all of the elements just + or - hence 0/1 knapsack variation

cache={}
def dp(i, sign, cur):
    if (i, sign, cur) in cache:
        return cache[(i, sign, cur)]

    if i>len(nums)-1:
        if cur==target:
            return 1/2
        return 0

    if sign=='+':
        res=dp(i+1, '+', cur+nums[i]) + dp(i+1, '-', cur+nums[i])
    if sign=='-':
        res=dp(i+1, '+', cur-nums[i]) + dp(i+1, '-', cur-nums[i])

    cache[(i, sign, cur)] = res

    return res

return int(dp(0, '+', 0) + dp(0, '-', 0))
```

```python
#interleaving string

#Given strings s1, s2, and s3, find whether s3 is formed by an interleaving of s1 and s2.
#An interleaving of two strings s and t is a configuration where s and t are divided into n and m substrings
#respectively, such that:
# - s = s1 + s2 + ... + sn
# - t = t1 + t2 + ... + tm
# - |n - m| <= 1
#The interleaving is s1 + t1 + s2 + t2 + s3 + t3 + ... or t1 + s1 + t2 + s2 + t3 + s3 + ...

#we can model this as a matrix with the matricies used in biology for allinging sequences
#however we modify this a bit, it can then become a path problem in this matix
#with s1 on the right and s2 on top and - at the 0 position of both we start at 0,0 and try to get to
#the bottom right square of the matrix. We can only move right or down and we only move to a square if 
#it mathces the corresponding position in the string s3
#this also repects the 3 condition given |n-m|<=1 since one continuous right arrow(made of sequence of right steps)
#would correspond to a ti for string s2. Likewise for a continuous down arrow and s1.
#we notice that evevery continous right arrow(except possibly the last) is followed by a down arrow(and vice versa)
#so the difference in number of coninuous right arrows and down arrows is 1 at most

cache={}
def dp(i, j, k):
    #base case
    if i>=len(s1)-1 and j>=len(s2)-1 and k>=len(s3)-1:
        return True
    if (i, j) in cache:
        return cache[(i, j)]

    res=False
    #print(i, j, k)
    if i+1<=len(s1)-1 and s1[i+1]==s3[k]:
        res = dp(i+1, j, k+1)
    if j+1<=len(s2)-1 and s2[j+1]==s3[k]:
        res = res or dp(i, j+1, k+1)

    cache[(i, j)] = res
    return res

if len(s1)+len(s2)!=len(s3):
    return False
return dp(-1, -1, 0)
```

```python
#longest increasing path in a matrix

#Given an m x n integers matrix, return the length of the longest increasing path in matrix.
#From each cell, you can either move in four directions: left, right, up, or down. 
#You may not move diagonally or move outside the boundary (i.e., wrap-around is not allowed).

#the solution uses the following recurrence:
#f(i, j) = max(1+f(i+1, j), 1+f(i-1, j), 1+f(i, j+1), 1+f(i, j-1))
#where arguments are only present if at that point in matrix, number is greater than at (i,j)
#note this solution works since we want a STRICTLY INCREASING path
#if it wasn't strictly increasing we could have problem with loops
#also f(i, j) denotes the length of the longest path starting at (i, j) in the matrix

dp=[[0 for j in range(0, len(matrix[0]))] for i in range(0, len(matrix))]

def f(i, j):
    if i>=len(matrix) or j>=len(matrix[0]):
        return 0
    if dp[i][j]!=0:
        return dp[i][j]

    up=-10**6
    down=-10**6     #default values that way they don't get picked in max
    left=-10**6     #if associated value in matrix is not greater(ie we do not follow that path)
    right=-10**6
    if i+1<len(matrix) and matrix[i+1][j]>matrix[i][j]:
        down=f(i+1, j)
    if i-1>=0 and matrix[i-1][j]>matrix[i][j]:
        up=f(i-1, j)
    if j+1<len(matrix[0]) and matrix[i][j+1]>matrix[i][j]:
        right=f(i, j+1)
    if j-1>=0 and matrix[i][j-1]>matrix[i][j]:
        left=f(i, j-1)

    dp[i][j] = 1 + max(0, up, down, left, right)   #put 0 also that way if its larger than all neighbours largest set
    return dp[i][j]                                #starting at it is 1

m=0
for i in range(0, len(matrix)):
    for j in range(0, len(matrix[0])):
        m = max(m, f(i, j))
return m
```

```python
#burst balloons

#You are given n balloons, indexed from 0 to n - 1. Each balloon is painted with a number on it represented by an array nums. 
#You are asked to burst all the balloons.
#If you burst the ith balloon, you will get nums[i - 1] * nums[i] * nums[i + 1] coins. If i - 1 or i + 1 goes out of bounds 
#of the array, then treat it as if there is a balloon with a 1 painted on it.
#Return the maximum coins you can collect by bursting the balloons wisely.
#
#example: nums=[3,1,5,8], coins =  3*1*5    +   3*5*8   +  1*3*8  + 1*8*1 = 167

#in the solution we must find a way to implement dp but if we consider the method of popping balloons first
#the subproblems are not independant so we cannot use dp.
#to use dp, we consider the subrpoblem of the maximum of coins if we pop that balloon last in a subarray of given length
#if that balloon is popped last for a subarray of given length then the problems of coins from balloons on right and left of 
#that subarray become independant since we know one will not affect the other
#we let f(i, j) be the maximum amount of coins we can get by popping balloons in the subarray from indicies i to j
#so the answer we want if f(0, len(nums)-1)
#for each balloon in the subarray of indicies i to j, find what the max is by popping balloon k last
#thus the recursion becomes
#f(i, j) = max(f(i, k-1) + f(k+1, j) + nums[i-1]*nums[k]*nums[j+1]) ,for all k in range i<=k<=j
#if we get case where k==i or k==j, those f's return 0
#these results are kept in dp table but only need to be filled on the upper diagonal
#since we only consider arrays from index i to index j we never need any case below the main diagonal since j<i in those cases


class Solution:
    def maxCoins(self, nums: List[int]) -> int:
        dp=[[0 for j in range(0, len(nums))] for i in range(0, len(nums))]

        n=len(nums)
        for l in range(1, n+1):
            for i in range(0, n-l+1):
                j=i+l-1
                for k in range(i, j+1):
                    if i==0:
                        leftNeigh=1
                    else:
                        leftNeigh=nums[i-1]
                    if j==n-1:
                        rightNeigh=1
                    else:
                        rightNeigh=nums[j+1]
                    val=leftNeigh*nums[k]*rightNeigh
                    
                    if k==i:
                        left=0
                    else:
                        left=dp[i][k-1]
                    if k==j:
                        right=0
                    else:
                        right=dp[k+1][j]
                    
                    dp[i][j]=max(dp[i][j], left+val+right)

        return dp[0][n-1]
```

```python
#regular expression matching

#you are given a string s and a pattern p
#implement regular expression matching with '.' and '*' where:
#  '.' matches a single occurence of any character
# '*' matches 0 or more occurences of the character below it
#
#return True is s matches expression p, False othersie
#example: s='aa', p='a*' returns True
#example: s='aa', p='.' returns False
#example: s='aa', p='a' returns False
#example: s='adsvweb', p='.*' returns True since '.*' means match any character 0 or more times
#example: s='agrigfdb', p='a.*c' returns False
#example: s='agrigfdbc', p='a.*c' returns True

#in the solution we use the method used to allign molecular sequences in biology with a twist on the method
#fistly we must consider every 'x*' as one single character in a string p, which is why we build a list l
#where l holds every character of p and if there is a 'x*' then it is considered as a single character in l
#after that we apply the biology method with s and the string given by l
#where dp[i][j] indicates if s1s2...si matches regular expression given by l1l2...lj

class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        #first things first:
        #we need a way to consider x* characters as one instead of two
        #so we will hold every char of p in a list where x* is considred one single char
        l=[]
        indiciesOfStars=set()
        for i in range(0, len(p)):
            if p[i]=='*':
                indiciesOfStars.add(i)
        i=0
        while i<len(p):
            if i+1 in indiciesOfStars:
                l.append(p[i]+p[i+1])
                i+=2
                continue
            l.append(p[i])
            i+=1
        #so now l is the string representation we want for p
        #because x* is considered as a single character
        
        #now we do a variation of biology algorithm of aligning two sequences
        dp=[[0 for j in range(0, len(l)+1)] for i in range(0, len(s)+1)]
        dp[0][0]=True

        for j in range(1, len(l)+1):
            if '*' in l[j-1] and dp[0][j-1]:
                dp[0][j]=True
            else:
                dp[0][j]=False
        for i in range(1, len(s)+1):
            dp[i][0]=False

        for i in range(1, len(s)+1):
            for j in range(1, len(l)+1):
                if '.*' == l[j-1]:
                    dp[i][j]=(dp[i-1][j]) or (dp[i][j-1]) or (True and dp[i-1][j-1])
                elif '*' in l[j-1] and s[i-1] in l[j-1]:
                    dp[i][j]=(dp[i-1][j]) or (dp[i][j-1]) or (True and dp[i-1][j-1])
                elif '*' in l[j-1]: #case where * in it but letter doesnt match
                    dp[i][j] = dp[i][j-1]
                elif '.' in l[j-1] or s[i-1] == l[j-1]:
                    dp[i][j] = True and dp[i-1][j-1]
                else:
                    dp[i][j] = False

        return dp[len(s)][len(l)]
```

```python
#jump game

#ou are given an integer array nums. You are initially positioned at 
#the array's first index, and each element in the array represents your maximum jump length at that position.
#Return true if you can reach the last index, or false otherwise.

#notice we can use a greedy solution in this case. We start from the last index and then move to the first element that can jump
#to goal and make that our new goal and try to find an element that can reach this new goal and pick the first that does
#notice the reason we can use greedy here i.e. the reason we can always pick the first by going in reverse order
#is that if there is another index further left from the first one that brings us to goal, then it can bring us to the first
#one that brings us to goal and then we can go from goal from there
#as an exmaple [1, 3, 4, 2] we move to 4 immediately, but 3 could also bring us to last index, this doesnt matter because 
#if 3 can bring us to last index, then it can bring us to 4 and then from 4 we can go to 2. This works because we are not
#looking for a solution containig the minimum number of jumps, we just want to know if there is a solution
#and this greedy approach let's us do this in O(n) time and O(1) memory which is very efficient

goal=len(nums)-1
for i in range(len(nums)-2, -1, -1):
    if nums[i] >= goal-i:
        goal=i
return goal==0
```

```python
#jump game II

#sliding window algorithm, for every jump, we calculate the range of indicies that that jump can take us at,
#right is the furthest, and left is the least furthest, this works since we are given that a solution exists

left=0
right=0
res=0
while right<len(nums)-1:
    furthest=0
    for i in range(left, right+1):
        furthest=max(furthest, i+nums[i])
    left=right+1
    right=furthest
    res+=1
return res
```

```python
#hand of straights

#Alice has some number of cards and she wants to rearrange the cards into groups so that each group is of size groupSize, 
#and consists of groupSize consecutive cards.
#Given an integer array hand where hand[i] is the value written on the ith card and an integer groupSize, 
#return true if she can rearrange the cards, or false otherwise.

#make a dictionary of counts of elements, also add each unique element to a min heap
#then on every turn, try to make a hand of length groupSize that starts with the min element left
#we can get the min element left by looking at the top of the heap
#if we cannot then return False immediately. Also, while we are making a hand,
#we decrement the count in the counts dictionary by 1 for the particular element we added to the hand
#if the count ever gets to 0, remove it from the dictionary and pop from the heap
#if the element popped from the heap(the min) is not the same as the element whos count just went to 0 then we know we
#have a problem since the current one is needed for making a hand including the min value, but since min value since in
#heap that means we still have some left, meaning we will make a another group with it in future but element that we 
#just ran out of will also be needed in future so return False

if len(hand)%groupSize!=0:
    return False
counts={}
for i in hand:
    if not i in counts:
        counts[i]=0
    counts[i]+=1
heap=list(counts.keys())
heapq.heapify(heap)
while len(heap)!=0:
    first=heap[0] #min value
    #try to create group of size groupSize starting with first
    for i in range(first, first+groupSize):
        if not i in counts:
            return False
        counts[i]-=1
        if counts[i]==0:
            if heap[0]!=i:
                return False
            del counts[i]
            heapq.heappop(heap)
return True
```

```python
#partition labels

#You are given a string s. We want to partition the string into as many parts as possible so that 
#each letter appears in at most one part.
#Note that the partition is done so that after concatenating all the parts in order, the resultant string should be s.
#Return a list of integers representing the size of these parts.
#
#example: s = "ababcbacadefegdehijhklij", Output: [9,7,8]
#example: s = "eccbbbbdec", Output: [10]

#the solution uses a sort of sliding window algorithm. we keep a hash map to contain the indicies of each element
#in reality only the largest index of that element is needed
#the we loop through the string, keeping track of the last index of the current substring with lastOfCur
#if we get to a point where i>lastOfCur, then we start a new substring

indicies={}
for i in range(0, len(s)):
    if not s[i] in indicies:
        indicies[s[i]]=[]
    indicies[s[i]].append(i)
res=[]
firstOfCur=0
lastOfCur=indicies[s[0]][-1]
for i in range(1, len(s)):
    if i>lastOfCur:
        res.append(lastOfCur-firstOfCur+1)
        firstOfCur=i
        lastOfCur=indicies[s[i]][-1]
        continue
    lastOfCur=max(lastOfCur, indicies[s[i]][-1])
res.append(lastOfCur-firstOfCur+1)
return res
```

```python
#valid parenthesis string

#given a string s consisting of only '(', ')', '*', where '*' could be replaced by '(', ')', or by the empty string
#return if s is a valid parenthesis string or not

#on the stacks we will push indicies of elements encountered
#stack1 will contain indicies for '(' character and will get popped 
#when a corresponding ')' character is encountered
#stack2 will contain indicies for '*' characters
#notice error only happens if:
#if we have extra '(' characters that arent matched,
#or if we have extra ')' that arent matched
#in the case of unmatched ')', we try to match them with * left, we can simply pop * from stack2
#in the case of unmatched '(', we try to match them with * that occur AFTER the '(' in question, this is why indicies of * and (
#are stored in order to see if a * will match with a (

stack1=[]
stack2=[]
for i in range(0, len(s)):
    char=s[i]
    if char=='(':
        stack1.append(i)
    if char==')':
        if len(stack1)!=0:
            stack1.pop()
        elif len(stack2)!=0:
            stack2.pop()
        else:
            return False
    if char=='*':
        stack2.append(i)
if len(stack1)==0:
    return True
i=0 #stack1 pointer
j=0 #stack2 pointer
while i<len(stack1):
    while j<len(stack2) and stack1[i]>stack2[j]:
        j+=1
    if j==len(stack2):  #if out of stars but still '(' left, return False
        return False
    #if we get here that means we found a j to associate to i
    #meaning we found a star with later index than once for '(' at i
    #and so we can match them and we move on to the next star
    #we also move to next unmatched '(', since this is a while loop we increment i
    j+=1
    i+=1
return True #if we can match all of them, then return True
```

```python
#insert interval

#You are given an array of non-overlapping intervals intervals where intervals[i] = [starti, endi] represent the start and the end of the ith interval and intervals is sorted in ascending order by starti. You are also given an interval newInterval = [start, end] that represents the start and end of another interval.
#Insert newInterval into intervals such that intervals is still sorted in ascending order by starti and intervals still does not have any overlapping intervals (merge overlapping intervals if necessary).
#Return intervals after the insertion.
#Note that you don't need to modify intervals in-place. You can make a new array and return it.
#example: Input: intervals = [[1,3],[6,9]], newInterval = [2,5], Output: [[1,5],[6,9]]
#example: Input: intervals = [[1,2],[3,5],[6,7],[8,10],[12,16]], newInterval = [4,8]Output: [[1,2],[3,10],[12,16]]

#in the solution we make a new array res which will be returned, this makes code simpler to avoid dealing with modifying
#intervals in place. We then calculate the new interval that will include this given newInterval
#this oneIncluding interval may end up being larger than given newInterval and takes care of the overlapping problem
#we also have a variable needToAddOneIncluding which indicates us when we need to add this calculated interval

res=[]
x=newInterval[0]
y=newInterval[1]
oneIncluding=[x, y]
needToAddOneIncluding=True
for lis in intervals:
    a=lis[0]
    b=lis[1]
    if a>y:    #in this case [x,y] appears entirely before [a,b]
        if needToAddOneIncluding:
            res.append(oneIncluding)
            needToAddOneIncluding=False
        res.append(lis)
        continue
    if a==y: #in this case we know interval including ends at b
        oneIncluding[1]=b #end of interval containing newInterval is b
    if a<y:     #in this case we must check
        if b<x:        #this means [a,b] occurs entirely before [x,y]
            res.append(lis)
            continue
        if b==x:      #this means interval including stars at a
            oneIncluding[0]=a
        #in this case a<y and b>x which means intervals do overlap
        #in any case, we get that left endpoint=min(a,x)
        #and right endpoint=max(b, y)
        #since we might have more than one interval overlapping with [x,y],
        #do not use x,y directly to update intervalIncluding
        #instead use the values stored in intervalIncluding
        oneIncluding[0]=min(oneIncluding[0], a)
        oneIncluding[1]=max(oneIncluding[1], b)
if needToAddOneIncluding:
    res.append(oneIncluding) 
return res
```

```python
#merge intervals

#Given an array of intervals where intervals[i] = [starti, endi], merge all overlapping intervals, 
#and return an array of the non-overlapping intervals that cover all the intervals in the input.

if len(intervals)==1:
    return intervals
intervals.sort(key = lambda x: x[0])
res=[]
i=0
j=1
while i<len(intervals) and j<len(intervals):
    x, y = intervals[i]
    a, b = intervals[j]
    #check if intervals [x,y] and [a, b] overlap
    #notice it is impossible so [a,b] to occur entirely before [x,y]
    #since intervals are sorted by their first component
    #so either [a,b] occurs entirely after [x,y] or [a,b] and [x,y] overlap
    if a>y: #in this case [a,b] occurs entirely after [x,y]
        res.append(intervals[i])
        i+=1
        j+=1
        continue
    #else intervals overlap, in any case the overlapping resulting interval
    #is [min(x,a), max(y, b)]
    newInterval=[min(x, a), max(y, b)]
    #this includes both intervals so make intervals[j] this new interval
    #and set i=j and j+=1
    intervals[j]=newInterval
    i=j
    j+=1
while i<len(intervals):
    res.append(intervals[i])
    i+=1 #in theory this line is executed at most once
return res
```

```python
#longest palindromic subsequence

#given a string s, find it's longest palindromic subsequence. len(s)>=1

#note in this question we are looking for subsequence not substring, ie the subsequence need not be contiguous in s
#but the relative order of the elements must be maintained. 
#To solve this, notice that if we take s and it's reverse, a common subsequence is necessairly a palindrome since
#it will be read the same in the reverse order as in the regular order.  Further more notice that any palindrome will be indeed 
#a common subsequence for that exact reason, a palindrome is read the same forward and backward
#so it will be a common subsequence of s and it's reverse. Now that we have established this,
#we can simply look for the longest common subsequence of s and it's reverse using the biology method

s2=s[::-1]
dp=[[0 for j in range(0, len(s2)+1)] for i in range(0, len(s)+1)]
for i in range(1, len(s)+1):
    for j in range(1, len(s2)+1):
        if s[i-1]==s2[j-1]:
            dp[i][j] = max(dp[i-1][j], dp[i][j-1], 1+dp[i-1][j-1])
        else:
            dp[i][j] = max(dp[i-1][j], dp[i][j-1])
return dp[len(s)][len(s2)]
```

```python
#maximum good subarray sum 

#ou are given an array nums of length n and a positive integer k. A subarray of nums is called good 
#if the absolute difference between its first and last element is exactly k, 
#in other words, the subarray nums[i..j] is good if |nums[i] - nums[j]| == k.
#Return the maximum sum of a good subarray of nums. If there are no good subarrays, return 0.

#in the solution we keep track of the prefix sum corresponding to every num in nums,
#then for every number x, we check if the corresponding number(x+k or x-k) is in our hashmap
#and subtract the current prefix sum with the prefix sum of the corresponding number
#note we always keep the minimum prefix sum of a corresponding number in the hash table
#in case we have duplicates since we want our result to be the largest
#so we'd subtract smallest in order to get largest result. 
#Also, we only check for corresponding numbers of x that have occured before x, if there are correponding numbers later
#in nums, then x will be the corresponding number for it and x is already in hash map

current=0
res=float('-inf')
mn={}
#mn[x] is the smallest prefix sum with the last number "x"(not including x)
for x in nums:
    if x in mn:
        mn[x]=min(mn[x], current)
    else:
        mn[x]=current
    current+=x

    for y in [x-k, x+k]:
        if y in mn:
            res=max(res, current-mn[y])

if res==float('-inf'):
    return 0
return res
```

```python
#delete nodes and return forest

#Given the root of a binary tree, each node in the tree has a distinct value.
#After deleting all nodes with a value in to_delete, we are left with a forest (a disjoint union of trees).
#Return the roots of the trees in the remaining forest. You may return the result in any order.

#we use a dfs, we first visit both the children and do actions on them before potentially deleting the parent
#the dfs function works as follows, it gets called on both the children. Then it checks if the current node needs to be deleted
#if it does, then we make both the children have no parents if children are not None and if the children do not get deleted 
#either. Since if the the child gets deleted as well, we do not want it returned in the forest, only children who are
#not deleted will be returned in the forest. We then check if the children have to be deleted, if the children have to be
#deleted, we set that corresponding child as None for the current node


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def delNodes(self, root: Optional[TreeNode], to_delete: List[int]) -> List[TreeNode]:
        toDel=set()
        for i in to_delete:
            toDel.add(i)
        noParent=set({root})

        def dfs(node):
            if node == None:
                return
            dfs(node.left)
            dfs(node.right)
            if node.val in toDel:
                if node.left!=None and node.left.val not in toDel:
                    noParent.add(node.left)
                if node.right!=None and node.right.val not in toDel:
                    noParent.add(node.right)
            if node.left!=None and node.left.val in toDel:
                node.left=None
            if node.right!=None and node.right.val in toDel:
                node.right=None

        dfs(root)
        if root.val in toDel:
            noParent.remove(root)
        return list(noParent)
```

```python
#odd even jump

#You are given an integer array arr. From some starting index, you can make a series of jumps. 
#The (1st, 3rd, 5th, ...) jumps in the series are called odd-numbered jumps, and the (2nd, 4th, 6th, ...) jumps in 
#the series are called even-numbered jumps. Note that the jumps are numbered, not the indices.
#You may jump forward from index i to index j (with i < j) in the following way:
#During odd-numbered jumps (i.e., jumps 1, 3, 5, ...), you jump to the index j such that arr[i] <= arr[j] 
#and arr[j] is the smallest possible value. If there are multiple such indices j, you can only jump to the smallest such index j
#During even-numbered jumps (i.e., jumps 2, 4, 6, ...), you jump to the index j such that arr[i] >= arr[j] 
#and arr[j] is the largest possible value. If there are multiple such indices j, you can only jump to the smallest such index j
#It may be the case that for some index i, there are no legal jumps.
#A starting index is good if, starting from that index, you can reach the end of the array (index arr.length - 1) 
#by jumping some number of times (possibly 0 or more than once).
#Return the number of good starting indices.

#in the solution we use dort of dp to keep track of for a particular index if we can get to the end whether we start with
#an odd jump or with an even jump. We iterate from the numbers from right to left, i.e. we start at the end
#each seen number(i.e. number that current could possible jump to) is kept in a sorted list(sorted list takes O(log(n)) to add
#an elem). In this sorted list we binary search using the difined methods lowerBound and upperBound to check if there is 
#a possible elem to the right of current that we could jump to depending if we start with an even or odd jump
#if there is such an elem then we get it's index using our hash map and we go to that index in dp and get the corresponding 
#value. notice the index can be set with every iteration of the for loop by going backwards since in any case(even or odd jump)
#we always want the smallest index possible if there is a possible place we can jump

from sortedcontainers import SortedList
class Solution:
    def oddEvenJumps(self, arr: List[int]) -> int:
        if len(arr)==1:
            return 1

        def lowerBound(arr, elem):
            #finds the smallest array item that has a value >= elem
            #returns -1 if such item doesn't exist
            left=0
            right=len(arr)-1
            ans=-1
            while left<=right:
                mid=(right+left)//2
                if arr[mid]==elem:
                    return arr[mid]
                if arr[mid]<elem:
                    left=mid+1
                if arr[mid]>elem:
                    ans=arr[mid]
                    right=mid-1
            return ans

        def upperBound(arr, elem):
            #finds the largest array item that is <=elem
            #returns -1 if such item doesn't exist
            left=0
            right=len(arr)-1
            ans=-1
            while left<=right:
                mid=(left+right)//2
                if arr[mid]==elem:
                    return arr[mid]
                if arr[mid]>elem:
                    right=mid-1
                if arr[mid]<elem:
                    ans=arr[mid]
                    left=mid+1
            return ans

        dp=[[False for j in range(0, len(arr))] for i in range(0, 2)]
        #dp[i][j]=1 if from index j we can make it to the end doing a i jump
        #and it's equal to 0 otherwise
        #we will use the convention that i=0 corresponds to starting with an even jump
        #and i=1 corresponds to doing starting with an odd jump from index j
        indiciesOfElems={}
        s=SortedList()
        dp[0][-1]=True
        dp[1][-1]=True
        s.add(arr[-1])
        indiciesOfElems[arr[-1]] = len(arr)-1
        for i in range(len(arr)-2, -1, -1):
            oddJumpNextVal=lowerBound(s, arr[i])
            evenJumpNextVal=upperBound(s, arr[i])
            if oddJumpNextVal!=-1:
                oddJumpNextIndex = indiciesOfElems[oddJumpNextVal]
                dp[1][i] = dp[0][oddJumpNextIndex]
            else:
                dp[1][i] = False
            if evenJumpNextVal!=-1:
                evenJumpNextIndex = indiciesOfElems[evenJumpNextVal]
                dp[0][i] = dp[1][evenJumpNextIndex]
            else:
                dp[0][i] = False
            indiciesOfElems[arr[i]]=i
            s.add(arr[i])

        res=0
        for i in range(0, len(dp[0])):
            if dp[1][i]:
                res+=1
        return res

```

```python
#graph connectivity with threshold

#We have n cities labeled from 1 to n. Two different cities with labels x and y are directly connected by a bidirectional road 
#if and only if x and y share a common divisor strictly greater than some threshold. 
#More formally, cities with labels x and y have a road between them if there exists an integer z such that 
#all of the following are true:
#x % z == 0,
#y % z == 0, and
#z > threshold.
#Given the two integers, n and threshold, and an array of queries, you must determine for each queries[i] = [ai, bi] 
#if cities ai and bi are connected directly or indirectly. (i.e. there is some path between them).
#Return an array answer, where answer.length == queries.length and answer[i] is true if for the ith query, 
#there is a path between ai and bi, or answer[i] is false if there is no path.

#in the solution, we use a disjoint set data structure to see which parts or the graph are connected
#in order to connect actual nodes to another, for any two nodes i,j we must see if their gcd is greater than threshold
#if it is then we call union(i, j) else we do not connect i and j.
#We can do this efficiently in the following manner: notice that since we need gcd(or at least one divisor) greater than
#threshold, then there are no nodes less or equal to threshold that will have connections
#then we can start looping throught the nodes greater than threshold, and add a connection for every on of their multiples 
#that is <=n. After doing this we will get the correct resulting graph. Then for each query we check if the representative
#of the two nodes is the same, if so we append True to res because they are connected, else we append False to res.

class Solution:
    def areConnected(self, n: int, threshold: int, queries: List[List[int]]) -> List[bool]:
        #we can use a disjoint set data structure to answer the queries
        parent = [i for i in range(0, n+1)] #1 indexing so up to n+1
        rank = [1 for i in range(0, n+1)] #technically size but we'll name it rank
        def findParent(u):
            if u==parent[u]:
                return u
            listToUpdate=[]
            while u!=parent[u]:
                listToUpdate.append(u)
                u=parent[u]
            for node in listToUpdate:
                parent[node]=u
            return u
        def union(u, v):
            ru=findParent(u)
            rv=findParent(v)
            if rv==ru:
                return
            if rank[ru] < rank[rv]:
                parent[ru] = rv
                rank[rv]+=rank[ru]
            else:
                parent[rv] = ru
                rank[ru]+=rank[rv]
        
        for i in range(threshold+1, n+1):
            j=1
            while i*j<=n:
                union(i, i*j)
                j+=1
        ans=[]
        for query in queries:
            u, v = query
            if findParent(u) == findParent(v):
                ans.append(True)
            else:
                ans.append(False)
        return ans
```

```python
#most stones removed with same row or column

#On a 2D plane, we place n stones at some integer coordinate points. Each coordinate point may have at most one stone.
#A stone can be removed if it shares either the same row or the same column as another stone that has not been removed.
#Given an array stones of length n where stones[i] = [xi, yi] represents the location of the ith stone, 
#return the largest possible number of stones that can be removed.

#we use the disjoint set data structure. Any stones sharing a row or a column(sharing x or y val) will be in the same set
#notice that for a given set, if we chose the order right, we can remove every stone of that set except for 1(the last 
#one remaining). So to solve this problem, we build the disjoing set, and call union on any stones that share x or y val.
#We want to quickly know if stone i and j share a x or y val without having to loop through the whole list
#since looping through the whole list will make the algorithms O(n^2). 
#then we get the number of disjoint sets by getting the number of unique representatives, and this means that 
#we can remove every stone except for these representatives. Thus we return len(stones)-numRepresentatives

parent = [i for i in range(0, len(stones))] #stone i correponds to index i
rank = [1 for i in range(0, len(stones))]

def find(u):
    if u==parent[u]:
        return u
    parent[u]=find(parent[u])
    return parent[u]

def union(u, v):
    ru=find(u)
    rv=find(v)
    if ru==rv:
        return
    if rank[ru] < rank[rv]:
        parent[ru] = rv
        rank[rv]+=rank[ru]
    else:
        parent[rv]=ru
        rank[ru]+=rank[rv]

rowsMap={}
columnsMap={}

for i in range(0, len(stones)):
    x, y = stones[i] #we use index cause dijoint set is kept by indices
    if x not in rowsMap:
        rowsMap[x] = i
    if y not in columnsMap:
        columnsMap[y] = i
    union(i, rowsMap[x])
    union(i, columnsMap[y])

#after building the disjoint set
#we want to get the number of unique parents(representatives)
#we can then remove every stone that isn't a parent
seenRepresentatives=set()
numRepresentatives=0
for node in range(0, len(stones)):
    if find(node) not in seenRepresentatives:
        seenRepresentatives.add(find(node))
        numRepresentatives+=1
return len(stones)-numRepresentatives
```

```python
#find maximum number of elements in a subset

#You are given an array of positive integers nums.
#You need to select a subset of nums which satisfies the following condition:
#You can place the selected elements in a 0-indexed array such that it follows the pattern: 
#[x, x2, x4, ..., xk/2, xk, xk/2, ..., x4, x2, x] (Note that k can be be any non-negative power of 2). 
#For example, [2, 4, 16, 4, 2] and [3, 9, 3] follow the pattern while [2, 4, 8, 4, 2] does not.
#Return the maximum number of elements in a subset that satisfies these conditions.

#solution is similar to longest chain of consecutive elements in an array, but now we need 2 elements to be present to move
#to next in chain which is square of current
#counts of all elements are put in a dictionary. We then check, if it is the largest element in sequence(the x^k term)
#if it is, then find all the rest of elements smaller and calculate length of sequence, if not then keep going till
#we get to largest and then calculate in order to get longest sequence possible
#there is one egde case we have to worry about, a sequence of 1's satisfies conditions, so to get length
#of chain consisting of 1's only we just get the count of 1's. If it is an odd number then that's the length of sequence
#if it is even, then subtract 1 to the count and that will be length of the sequence
#at the end the length of the longest sequence found is returned

counts={}
for i in nums:
    if i not in counts:
        counts[i]=0
    counts[i]+=1
res=1
for i in counts:
    if (i**2 in counts and counts[i]>1) or i==1: #this line skips 1's as well
        continue
    #otherwise i is the top(x**k) elem of a subset
    #explore length of that subset
    cur=1
    while i**(1/2) in counts and counts[i**(1/2)]>1:
        i=i**(1/2)
        cur+=2
    res=max(res, cur)
#handle the edge case where there are 1's, since [1, 1, 1] is a valid
#subset for example
if 1 in counts:
    if counts[1]%2==0:
        res=max(counts[1]-1, res)
    else:
        res=max(res, counts[1])

return res
```

```python
#word subsets

#You are given two string arrays words1 and words2.
#A string b is a subset of string a if every letter in b occurs in a including multiplicity.
#For example, "wrr" is a subset of "warrior" but is not a subset of "world".
#A string a from words1 is universal if for every string b in words2, b is a subset of a.
#Return an array of all the universal strings in words1. You may return the answer in any order

#the naive way to implement this would be to get a counter for every word in words2
#and then for each word in words1 get a counter for it and check it with each counter of words2
#then keep a flag to check if it is universal or not
#the two nested for loops would make this algorithm inefficient
#the better way would be to notice that for a word from words1 to be universal, then EVERY string of words2
#must be a subset of it, but notice there may be some overlap in strings from word2
#for example if words2=['apple', 'l'] then the l is kind of useless, since it's contained in apple
#so any string for who apple is a subset is universal, but if words2=['apple', 'll'] then that second l changes things
#and now any strign for who applel is a subset is universal, so we sort of need to get the union of all strings in words2
#(considering multiplicity of individual strings) and then we could make 1 single counter for this union word
#and for every word in words1 we compare their counter with this counter of union not with alot of dictionaries
#this saves work since we are given that len(words2)<=10^4 so instead of comparing 10^4 dictionaries, we only compare one,
#of course this new union dicionary may be much larger since it is a dictioanry comprising the union of 10^4 words
#but as we saw with the example ['apple', 'l'], when we have 10^4 words each of length at most 10, there is alot of potential
#for words sharing letters so alot of repeated work will be eliminated

res=[]
d={}
for b in words2:
    counter={}
    for letter in b:
        if letter not in counter:
            counter[letter]=0
        counter[letter]+=1

    for key in counter:
        if key not in d:
            d[key]=counter[key]
        elif key in d and d[key]<counter[key]:
            d[key]=counter[key]

for a in words1:
    counter={}
    for letter in a:
        if letter not in counter:
            counter[letter]=0
        counter[letter]+=1
    toAdd=True
    for key in d:
        if key not in counter or d[key]>counter[key]:
            toAdd=False
            break
    if toAdd:
        res.append(a)

return res
```
